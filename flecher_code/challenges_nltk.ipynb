# Challenge 1

```python
import nltk
from textblob import TextBlob
from nltk.corpus import movie_reviews

fileids = movie_reviews.fileids()[:100]
doc_words = [movie_reviews.words(fileid) for fileid in fileids]
documents = [' '.join(words) for words in doc_words]
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1,2))
doc_vectors = vectorizer.fit_transform(documents)
features = vectorizer.get_feature_names()
```

```python
vectorizer1 = TfidfVectorizer(stop_words="english", ngram_range=(1,2),max_features=10)
doc_vectors1 = vectorizer1.fit_transform(documents)
features1 = vectorizer1.get_feature_names()
```

```python
features1
```

```python
for doc in doc_vectors:
    indxsort = doc.toarray()[0].argsort()[-10:][::-1]
    print [features[i] for i in indxsort]
    print
```

# Challenge 2

```python
from pymongo import MongoClient
client = MongoClient()
congress = client.dsbc.congress
```

```python
tweet_cursor = congress.find({},{'text'})
tweets = [t['text'] for t in tweet_cursor]
```

```python
vectorizer_t = TfidfVectorizer(stop_words="english", ngram_range=(1,2),max_features=10)
tweet_vectors = vectorizer_t.fit_transform(tweets)
tweet_features = vectorizer_t.get_feature_names()
```

```python
tweet_features
```

```python
for t in tweet_vectors:
    indxsort = t.toarray()[0].argsort()[-5:][::-1]
    print [tweet_features[i] for i in indxsort]
    print
```
